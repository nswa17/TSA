関数化

```julia
function get_taisu_sabun(xs)
    return Float64[log(xs[i+1]) - log(xs[i]) for i in 1:(length(xs)-1)]
end
```

```julia
plot(get_taisu_sabun(df[:exrate]))
```

```julia
plot(get_taisu_sabun(df[:indprod]))
```

```julia
using StatsBase
```

```julia
autocor(collect(df[:indprod]), [1, 2, 3, 4, 5])
```

```julia
autocor(get_taisu_sabun(df[:indprod]), [0, 1, 2, 3, 4, 5])
```

```julia
function Q(m, T, data)
    rhos = autocor(get_taisu_sabun(data), collect(1:m))
    return T*(T+2)*sum([rho^2/float(T-k) for (k, rho) in enumerate(rhos)])
end
```

```julia
Q(m) = Q(m, length(df[:indprod]), df[:indprod])
```

```julia
[Q(i) for i in 1:50]
```

```julia
bar(collect(1:20), autocor(get_taisu_sabun(df[:indprod]), collect(1:20)),color="#0f87bf",align="center",alpha=0.4)
```
